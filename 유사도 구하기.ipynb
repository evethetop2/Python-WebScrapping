```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('아이유전곡최종.csv', encoding='cp949')
df.head()
```

```python
# 데이터 정보 확인하기
df.info()
```

```python
df.drop_duplicates(subset = 'title', keep='first',inplace=True)
```

```python
df.reset_index(drop=True, inplace= True)
df
```

```python
# 간단한 명사 단어 전처리
from konlpy.tag import Okt
okt = Okt()
```

```python
# 명사 단어 전처리
df['Lyric'].map(lambda x :okt.nouns(x))
```

```python
df['Lyric'] = df['Lyric'].map(lambda x: ' '.join(x)) #transform 은 lower가 있는데 리스트는 불가능
```

```python
df[df['Lyric'].map(lambda x : '가게 샴푸' in x)]
```

```python
# TF-IDF 벡터화
vectorizer = TfidfVectorizer(min_df= 1, ngram_range=(1,1))
features = vectorizer.fit_transform(df['Lyric'])
```

```python
# feature 이름 불러 오기
feature_names = vectorizer.get_feature_names_out()
```

```python
# DTM 생성
dtm_np = np.array(features.todense())
```

```python
# DataFrame 확인
pd.DataFrame(data = dtm_np, columns = feature_names)
```

```python
# 코사인 유사도 구하기
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(dtm_np, dtm_np)
```

```python
pd.DataFrame(cosine_sim)
```

```python
# 도서명 기준 인덱스 설정 인덱스
indices = pd.Series(df.index, index=df['title'])
```

```python
# 도서명 확인 하기
df['title']
```

```python
# 마음에 드는 책 인덱스 추출
idx = indices['Celebrity']
print(idx)
```

```python
# 선택한 책과 다른 책의 유사도 확인
sim_scores = list(enumerate(cosine_sim[idx]))
sim_scores
```

```python
# 유사도가 높은 순으로 책 정렬
sim_scores = sorted(sim_scores, key= lambda x: x[1], reverse= True)
```

```python
# 유사도 높은 3개의 책
sim_scores = sim_scores[1:4]
```

```python
# 가장 유사한 3개의 책의 인덱스
book_indices = [i[0] for i in sim_scores]
```

```python
df['title'].iloc[book_indices].tolist()
```

```python
# 유사한 책 제목 출력
title = 'Celebrity'
data = title, df['title'].iloc[book_indices].tolist()
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    # 선택한 책의 인덱스를 가지고 옵니다.
    idx = indices[title]

    # 검색한 책과 모든 책의 유사도를 구합니다.
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 유사도가 높은 순으로 책을 정렬합니다.
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 가장 유사한 4개의 책을 가지고 옵니다.
    sim_scores = sim_scores[1:4]

    # 가장 유사한 4개의 책의 인덱스를 받아옵니다.
    book_indices = [i[0] for i in sim_scores]

    # 가장 유사한 10개의 책의 제목을 반환합니다.
    return df['title'].iloc[book_indices].tolist() # 리스트 형태로 반환해야 인덱스 값이 제외됩니다.
```

```python
get_recommendations('Celebrity')
```

```python
df['title'].map(get_recommendations) # .map(lambda x : get_recommendations(x))와 동일합니다.
```

```python
# zip 함수는 두 리스트를 한 쌍으로 묶어줍니다. 여기서는 제목과 유사곡이 되겠죠!
sim_list = list(zip(df['title'], df['title'].map(get_recommendations)))
```

```python
sim_df = pd.DataFrame(sim_list, columns = ['제목', '유사곡'])
```

```python
sim_df.explode('유사곡') # 리스트를 폭파시킵니다!
```
